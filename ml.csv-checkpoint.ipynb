```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
```

```python
df=pd.read_csv("Cleaned_data.csv")
```

```python

X = df.iloc[:, :-1]  # Features (all columns except the last one)
y = df.iloc[:, -1]   # Target variable (last column)

# Split the dataset into 80% training and 20% testing, with random_state for reproducibility
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=44, shuffle=True)

# Print the shapes of the training and testing sets
print("Training set shape:", X_train.shape, y_train.shape)
print("Testing set shape:", X_test.shape, y_test.shape)
```

```python
X_train
```

```python
nbc=GaussianNB()
nbc.fit(X_train,y_train)


```

```python
# Make predictions on the testing data
y1_pred =nbc.predict(X_test)
print(classification_report(y_test, y1_pred))
print(confusion_matrix(y_test, y1_pred))
```

```python
# Evaluate the accuracy of the classifier
accuracy1 = accuracy_score(y_test, y1_pred)
print("Accuracy:", accuracy1)
```

```python
# Create an instance of the model
dtc = DecisionTreeClassifier()

# Train the model
dtc.fit(X_train, y_train)
```

```python

# Make predictions
y2_pred = dtc.predict(X_test)
print(classification_report(y_test, y2_pred))
print(confusion_matrix(y_test, y1_pred))
```

```python
# Evaluate the model
accuracy2 = accuracy_score(y_test, y2_pred)
print(f'Accuracy: {accuracy2}')
```

```python
knn = KNeighborsClassifier(n_neighbors=6)  
knn.fit(X_train, y_train)

```

```python
y3_pred = knn.predict(X_test)
print(classification_report(y_test, y3_pred))
print(confusion_matrix(y_test, y1_pred))
```

```python
accuracy3 = accuracy_score(y_test, y3_pred)
print(f'Accuracy: {accuracy3}')
```

```python
results = pd.DataFrame({'Algorithm': ['Naive Bayes', 'Decision Tree', 'KNN'],
                       'Accuracy': [ 0.8612984822934233,0.8739460370994941, 0.8490725126475548]})
```

```python
plt.figure(figsize=(5, 3))
bar_width=0.1
plt.bar(results['Algorithm'], results['Accuracy'])
plt.xlabel('Algorithm')
plt.ylabel('Accuracy')
plt.title('Comparison of Algorithm Accuracies')
plt.show()
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
```

```python
table = pd.DataFrame(y_pred, columns=['Predicted Values'])
table.index = y_test

table
```

```python
rmsex = np.sqrt(mean_squared_error(y_test, y_pred))
maex = mean_absolute_error(y_test, y_pred)
print(f'linear regression RMSE: {rmsex}')
print(f'linear regression MAE: {maex}')
```

```python
knnr = KNeighborsRegressor(n_neighbors=6)  

knnr.fit(X_train, y_train)

y_pred = knnr.predict(X_test)
```

```python
table = pd.DataFrame(y_pred, columns=['Predicted Values'])
table.index = y_test

table
```

```python
rmsey = np.sqrt(mean_squared_error(y_test, y_pred))
maey = mean_absolute_error(y_test, y_pred)
print(f'KNN Regressor RMSE: {rmsey}')
print(f'KNN Regressor MAE: {maey}')
```

```python
dtr = DecisionTreeRegressor()
dtr.fit(X_train, y_train)

y_pred = dtr.predict(X_test)
```

```python
table = pd.DataFrame(y_pred, columns=['Predicted Values'])
table.index = y_test

table
```

```python
rmsez = np.sqrt(mean_squared_error(y_test, y_pred))
maez = mean_absolute_error(y_test, y_pred)
print(f'Decision Tree Regressor RMSE: {rmsez}')
print(f'Decision Tree Regressor MAE: {maez}')
```

```python
results = pd.DataFrame({'Algorithm': ['Linear Regression',  'Decision Tree', 'KNN'],
                       'RMSE': [0.278978052369896, 0.3514604557262335, 0.3636926504804444],
                       'MAE': [ 0.17463887309908613, 0.12352445193929174, 0.221964586846543]})
```

```python
# Plot the error metrics
plt.figure(figsize=(5, 3))
bar_width = 0.35
index = np.arange(len(results['Algorithm']))

plt.bar(index, results['RMSE'], width=bar_width, label='RMSE')
plt.bar(index + bar_width, results['MAE'], width=bar_width, label='MAE')

plt.xlabel('Algorithm')
plt.ylabel('Error')
plt.title('Comparison of Algorithm Performance (RMSE and MAE)')
plt.xticks(index + bar_width/2, results['Algorithm'])
plt.legend()
plt.tight_layout()
plt.show()
```
